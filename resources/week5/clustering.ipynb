```python
import pandas
import sklearn.datasets
```

```python
flowers = sklearn.datasets.load_iris()
```

```python
import sklearn.cluster
```

```python
clusterer = sklearn.cluster.KMeans(n_clusters=3)
```

```python
cluster_suggestions = clusterer.fit_predict(flowers['data'])
cluster_suggestions
```

```python
import sklearn.metrics
```

```python
sklearn.metrics.silhouette_score(flowers['data'], cluster_suggestions)
```

```python
clusterer10 = sklearn.cluster.KMeans(n_clusters=10)
cluster_suggestions10 = clusterer10.fit_predict(flowers['data'])
sklearn.metrics.silhouette_score(flowers['data'], cluster_suggestions10)
```

```python
scores = []
for i in range(2, 21):
    clustereri = sklearn.cluster.KMeans(n_clusters=i)
    cluster_suggestionsi = clustereri.fit_predict(flowers['data'])
    scores.append(sklearn.metrics.silhouette_score(flowers['data'], cluster_suggestionsi))
scores
```

```python
%matplotlib inline
import matplotlib
matplotlib.pyplot.plot(range(2,21), scores)
```

# Supervised learning using clusters

```python
titanic_raw = pandas.read_csv('../../data/titanic.csv')
titanic = titanic_raw[titanic_raw.Age.notnull()].copy()
titanic['ScaledAge'] = titanic.Age / 20.0
titanic['Gender'] = numpy.where(titanic.Sex == 'male', 1, 0)
```

```python
import sklearn.neighbors
knn = sklearn.neighbors.KNeighborsClassifier(n_neighbors=5)
knn.fit(titanic[['Pclass', 'ScaledAge', 'Gender']], titanic.Survived)
```

```python
knn.predict( [ [1, 1.0, 0  ]])
```

```python
knn.predict(  [ [3 , 0.2, 1]])
```

```python
sacramento = pandas.read_csv('../../data/Sacramento-real-estate-transactions.csv')
```

```python
sacramento
```

```python
knr = sklearn.neighbors.KNeighborsRegressor(n_neighbors=5)
import sklearn.cross_validation
(trainX, testX, trainY, testY) = sklearn.cross_validation.train_test_split(
    sacramento[['latitude', 'longitude', 'beds']], sacramento.price
)    
```

```python
knr.fit(trainX, trainY)
```

```python
predictions = knr.predict(testX)
```

```python
matplotlib.pyplot.scatter(testY, predictions)
```

```python
matplotlib.pyplot.scatter(testY, predictions,
                          cmap=matplotlib.pyplot.cm.rainbow,
                          c=testX.beds)
```
