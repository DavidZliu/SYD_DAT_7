```python
sacramento = pandas.read_csv('../../data/Sacramento-real-estate-transactions.csv')
```

```python
sacramento.beds.plot.hist()
```

```python
sacramento.plot.scatter('longitude', 'latitude')
```

```python
def bed_colours(x):
    if x == 1: return 'red'
    if x == 2: return 'green'
    if x == 3: return 'blue'
    return 'pink'
```

```python
sacramento.plot.scatter('longitude', 'latitude', c=sacramento.beds.map(bed_colours))
```

```python
sacramento.plot.scatter('longitude', 'latitude', c=sacramento.baths,
                        colormap=matplotlib.pyplot.cm.rainbow)
```

```python
sydtrains = pandas.read_csv('../week1/sydtrains.csv', index_col=1, thousands=',')
```

```python
central = sydtrains.loc['Central'].set_index('YEAR').sort_index()
townhall = sydtrains.loc['Town Hall'].set_index("YEAR").sort_index()
wynyard = sydtrains.loc['Wynyard'].set_index("YEAR").sort_index()
```

```python
import matplotlib.pyplot
(fig, ax) = matplotlib.pyplot.subplots()
ax.plot(central.IN_24_HOURS, color='r', label="Central")
ax.plot(townhall.IN_24_HOURS, color='g', label='Town Hall')
ax.plot(wynyard.IN_24_HOURS, color='b', label='Wynyard')
ax.legend(loc='best')
ax.set_title("Number of passengers catching a train")
```

```python

(fig, ax) = matplotlib.pyplot.subplots(nrows=3, ncols=1, sharex=True, sharey=True)
ax[0].plot(central.IN_24_HOURS, color='r', label="Central")
ax[1].plot(townhall.IN_24_HOURS, color='g', label='Town Hall')
ax[2].plot(wynyard.IN_24_HOURS, color='b', label='Wynyard')
```
