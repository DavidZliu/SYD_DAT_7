```python
import pandas
```

```python
movies = pandas.read_csv('../data/movies.csv', sep='|', 
                         usecols=[0,1], names=['movie_id', 'title']
                        )
movies
```

```python
ratings = pandas.read_csv('../data/movie-ratings.txt', sep='\t',
                         names=['user_id', 'movie_id', 'rating', 'timestamp'])
ratings
```

```python
one_slope_df = pandas.DataFrame({
        'title' : movies.set_index('movie_id').title,
        'ratings_counts': ratings.groupby('movie_id').rating.count(),
        'mean_rating': ratings.groupby('movie_id').rating.mean()
        })
one_slope_df
```

```python
one_slope_df.loc[one_slope_df[one_slope_df.ratings_counts > 25].mean_rating.idxmax()]
```

```python
ratings[ratings.movie_id != 408].user_id.value_counts()
```
